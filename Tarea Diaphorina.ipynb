el eje horizontal esté el mes y el vertical el promedio de diaphorinas por trampa.

Hacer una tabla donde la primera columna sea el mes, la segunda el índice

```python
meses=["Enero","Febrero","Marzo","Abril","Mayo","Junio","Julio",
        "Agosto","Septiembre","Octubre","Noviembre","Diciembre"]
#Creamos un diccionario donde a cada mes le asocia un dataframe con los datos de ese mes
data_por_mes = {}
i=1.0
for mes in meses:
    data_por_mes[mes] = data2015[data2015["Mes"]==i]
    i=i+1
```

```python
data_por_mes["Enero"].head()
```

```python
data_por_mes["Junio"].head()
```

```python
meses=["Enero","Febrero","Marzo","Abril","Mayo","Junio","Julio",
        "Agosto","Septiembre","Octubre","Noviembre","Diciembre"]
#Hacemos el diccionario con los group by, donde a cada mes le asocia la agrupación por promedio de trampa
group_por_mes = {}
for mes in meses:
    group_por_mes[mes] = data_por_mes[mes].groupby('Ubicacion')['No_DiaphorinasXtrampa'].mean() 
#Veamos un ejemplo
group_por_mes["Junio"]
```

```python
#Ahora para sacar el Id hacemos 
meses=["Enero","Febrero","Marzo","Abril","Mayo","Junio","Julio",
        "Agosto","Septiembre","Octubre","Noviembre","Diciembre"]
Id_por_mes = {}
for mes in meses:
   Id_por_mes[mes] = group_por_mes[mes].var()/ group_por_mes[mes].mean()
```

```python
Id_por_mes
```

```python
#Ahora para sacar el Cx hacemos 
meses=["Enero","Febrero","Marzo","Abril","Mayo","Junio","Julio",
        "Agosto","Septiembre","Octubre","Noviembre","Diciembre"]
Cx_por_mes = {}
for mes in meses:
   Cx_por_mes[mes] = (Id_por_mes[mes]-1)/(group_por_mes[mes].sum()-1)
```

```python
Cx_por_mes
```

```python
df = pd.DataFrame({
    'Mes': meses,
    'Id por mes': Id_por_mes.values(),
    'Cx por mes': Cx_por_mes.values()
})
df.index = range(1, len(df) + 1)
df
```
